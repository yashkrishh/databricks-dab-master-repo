# Default notebook

This default notebook is executed using a Lakeflow job as defined in resources/sample_job.job.yml.

```python
# Set default catalog and schema
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")
```

```python
# 🚗 Capturing the taxi data stream inside the 'df' variable box
df = spark.read.table("samples.nyctaxi.trips")
```

```python
# 1. Grab the dynamic distance filter from the YAML bundle configuration
# (We convert it to a float number so Python can calculate it)
max_distance = float(dbutils.widgets.get("notebook_param"))
print(f"Active Pipeline Threshold: Processing trips shorter than {max_distance} miles.")

# 2. Filter your live DataFrame on the fly using that parameter!
# (Assuming your DataFrame variable is named df)
filtered_df = df.filter(df.trip_distance < max_distance)
display(filtered_df)
```

```python
import os

# 🔌 Extraction Method 1: Reading from the Spark Configuration ledger
spark_param = spark.conf.get("spark.custom.partition_date")
print(f"Extraction Step 1 - Spark Config variable parsed: {spark_param}")

# 🔌 Extraction Method 2: Reading from the Linux Operating System container
env_param = os.getenv("PARTITION_DATE")
print(f"Extraction Step 2 - OS Environment variable parsed: {env_param}")
```
